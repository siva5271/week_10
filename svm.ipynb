# SVM classification using iris datasets

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn import svm
```

### Here i am using the iris datasets from sklearn to produce the necessary dataframes

```python
iris=load_iris()
df=pd.DataFrame(data=iris.data,columns=iris.feature_names)
df["target"]=iris.target
df["target names"]=df.target.apply(lambda x:iris.target_names[x])
df
```

## I am gonna plot the scatter plot to determine whether the data sets can be efficiently classified

```python
df0=df[df['target']==0]
df1=df[df['target']==1]
df2=df[df['target']==2]
df2
```

## setosa v/s versicolor

### sepal length and sepal width

```python
sns.scatterplot(x="sepal length (cm)",y="sepal width (cm)",data=df0)
sns.scatterplot(x="sepal length (cm)",y="sepal width (cm)",data=df1)
```

### petal length and petal width

```python
sns.scatterplot(x="petal length (cm)",y="petal width (cm)",data=df0)
sns.scatterplot(x="petal length (cm)",y="petal width (cm)",data=df1)
```

```python
X=df.drop(["target","target names"],axis=1)
y=df["target"]
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
model=svm.SVC()
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```
